# Data Collection and Preprocessing

```python
%pylab inline
import matplotlib.pyplot as pl
import pandas as pd
import pandas_profiling

# Import the dataset
data = pd.read_csv('boston-crime-raw.csv', names=['Incident Number', 'Offense Code', 'Offense Type',
                                              'Offense Description', 'District', 'Reporting Area',
                                              'Shooting', 'Date', 'Year',
                                              'Month', 'Day of Week', 'Hour',
                                              'UCR Offense Level', 'Street', 'Latitude',
                                              'Longitude', 'Coordinates'], skiprows=1, low_memory=False)
```

```python
data.head()
```

```python
# Reference: https://github.com/pandas-profiling/pandas-profiling
# Using Pandas Profiling to generate a compact summary of the dataset
pandas_profiling.ProfileReport(data)
```

From the above profile report we can see that there are 6.8% missing values that we need to take care of. The features that have missing values are Latitude, Longitude, Shooting and Street. There are also 721 duplicate rows which should be removed.

## Data cleaning and preprocessing

```python
# Remove unwanted features
data = data.drop(columns=['Offense Code'])
```

```python
# Add desired features
import numpy as np
data['Is Dark'] = 0
data.loc[(data['Hour'] >= 18), 'Is Dark'] = 1
data.loc[(data['Hour'] <= 5), 'Is Dark'] = 1
```

```python
# Replace district codes with district names
district_codes = pd.read_csv('district-codes.csv')
data['District'] = data['District'].replace(dict(zip(district_codes.DISTRICT, district_codes.DISTRICT_NAME)))
data['District'].unique()
```

```python
# Handle the missing data for Shooting by replacing the NaN values with 0
data['Shooting'] = data['Shooting'].fillna(0)
data['Shooting'].unique()
```

```python
# Correct the labels for Shooting from Y to 1
data['Shooting'] = data['Shooting'].replace('Y', 1)
data['Shooting'].unique()
```

```python
# Fill the null values for Latitude and Longitude with 0
data[['Latitude', 'Longitude']] = data[['Latitude', 'Longitude']].fillna(0)
data[['Latitude', 'Longitude']].isnull().any()
```

```python
# Remove duplicate rows
data = data.drop_duplicates()
data.duplicated().any()
```

```python
pandas_profiling.ProfileReport(data)
```

### Export the processed and cleaned dataset for further use

```python
data.to_csv('boston-crime.csv', index=False)
```
