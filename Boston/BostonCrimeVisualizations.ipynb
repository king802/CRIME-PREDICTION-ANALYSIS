 # Boston Crime Visualizations

```python
import numpy as np 
import pandas as pd 
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
# Use ggplot style for plots
plt.style.use('ggplot')
```

```python
# Import the dataset
df = pd.read_csv('boston-crime.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(15,10))

df.groupby([df['District']]).size().plot(kind='barh')
plt.title('Number of Crimes per District')
```

```python
plt.figure(figsize=(15,20))

df.groupby([df['Offense Type']]).size().sort_values().plot(kind='barh')
plt.title('Number of Crimes per Offense Type')
```

```python
plt.figure(figsize=(15,10))

df.groupby([df['Street']]).size().sort_values().tail(20).plot(kind='barh')
plt.title('Top 20 Most Unsafe Streets')
```

```python
plt.figure(figsize=(15,10))

df.groupby([df['Hour']]).size().plot(kind='bar')
plt.title('Number of Crimes per Hour')
```

```python
plt.figure(figsize=(15,10))

df.groupby([df['Shooting']]).size().plot(kind='bar')
plt.title('Number of Crimes involving Shooting')
```

```python
plt.figure(figsize=(15,10))
df.resample('M').size().plot(legend=False)

plt.ylabel('Crime Rate')
plt.xlabel('Months')
plt.title('Number of Crimes over the time period by Months')
```

```python
plt.figure(figsize=(15,10))
df.resample('Y').size().plot(legend=False)

plt.ylabel('Crime Rate')
plt.xlabel('Months')
plt.title('Number of Crimes over the time period by Year')
```

```python
primary_crimes_date = df.pivot_table('Incident Number', aggfunc=np.size, columns='Offense Type',index=df.index.date,fill_value=0)

plo = primary_crimes_date.rolling(365).sum().plot(figsize=(40, 70), subplots=True, layout=(-1,4), sharex=False)
```

```python
df.groupby([df.index.dayofweek]).size().plot(kind='barh')
plt.yticks(np.arange(7), ['Mon','Tue','Wed','Thur','Fri','Sat','Sun'])

plt.title('Crime Instances Per Day')
```

```python
is_2018 = df[df.Year == 2018]

plt.figure(figsize=(15,10))
is_2018.resample('M').size().plot(legend=False)

plt.ylabel('Crime Rate')
plt.xlabel('Months')
plt.title('Number of Crimes over the time period by Months in 2018')
```
